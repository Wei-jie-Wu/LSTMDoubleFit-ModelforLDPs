# Tanimoto（新颖性）

```python
import pandas as pd
import random
from rdkit import Chem
from rdkit.Chem import AllChem, DataStructs
import matplotlib.pyplot as plt
from tqdm import tqdm
import warnings
from rdkit import RDLogger
RDLogger.DisableLog('rdApp.*')

tqdm.pandas()  

# 读取数据
train_df = pd.read_csv("Modeldata0406.csv")
gen_df = pd.read_excel("NewMolecules0408.xlsx")

# 读取三个生成分子数据文件
opt_df_80 = pd.read_csv("80_90similar_newSMILES0501.csv")
opt_df_70 = pd.read_csv("70_80similar_newSMILES0501.csv")
opt_df_90 = pd.read_csv("90_100similar_newSMILES0501.csv")

# 随机抽样训练集 10 万个
train_smiles = train_df["SMILES"].dropna().unique().tolist()
train_sample = random.sample(train_smiles, 100000)

gen_smiles = gen_df["SMILES"].dropna().unique().tolist()
opt_smiles_80 = opt_df_80["SMILES"].dropna().unique().tolist()
opt_smiles_70 = opt_df_70["SMILES"].dropna().unique().tolist()
opt_smiles_90 = opt_df_90["SMILES"].dropna().unique().tolist()

def mol_from_smiles(smiles):
    try:
        return Chem.MolFromSmiles(smiles)
    except:
        return None

def get_fp(mol, fp_type="rdkit"):
    if mol is None:
        return None
    if fp_type == "rdkit":
        return Chem.RDKFingerprint(mol)
    elif fp_type == "morgan":
        return AllChem.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=2048)
    else:
        raise ValueError("Unknown fingerprint type")

ref_mols = [mol_from_smiles(smi) for smi in train_sample]
ref_rdkit_fps = [get_fp(mol, "rdkit") for mol in ref_mols if mol is not None]
ref_morgan_fps = [get_fp(mol, "morgan") for mol in ref_mols if mol is not None]

def calc_max_tc(smi_list, ref_fps, fp_type):
    results = []
    for smi in tqdm(smi_list):
        mol = mol_from_smiles(smi)
        if mol is None:
            continue
        fp = get_fp(mol, fp_type)
        if fp is None:
            continue
        sims = DataStructs.BulkTanimotoSimilarity(fp, ref_fps)
        results.append(max(sims))
    return results

# 计算Tanimoto系数
gen_rdkit = calc_max_tc(gen_smiles, ref_rdkit_fps, "rdkit")
opt_rdkit_80 = calc_max_tc(opt_smiles_80, ref_rdkit_fps, "rdkit")
opt_rdkit_70 = calc_max_tc(opt_smiles_70, ref_rdkit_fps, "rdkit")
opt_rdkit_90 = calc_max_tc(opt_smiles_90, ref_rdkit_fps, "rdkit")

gen_morgan = calc_max_tc(gen_smiles, ref_morgan_fps, "morgan")
opt_morgan_80 = calc_max_tc(opt_smiles_80, ref_morgan_fps, "morgan")
opt_morgan_70 = calc_max_tc(opt_smiles_70, ref_morgan_fps, "morgan")
opt_morgan_90 = calc_max_tc(opt_smiles_90, ref_morgan_fps, "morgan")


```

```python
def plot_histogram(data_dict, threshold, fp_type):
    import numpy as np
    plt.figure(figsize=(8, 6))

    bins = np.linspace(0, 1.0, 31) 
    colors = {
    "Generated": "#D3EEF9FF",
    "Optimized 70": "#A0FFFEFF",
    "Optimized 80": "#70B7E1FF",
    "Optimized 90": "hotpink"
    }

    for label, data in data_dict.items():
        plt.hist(data, bins=bins, alpha=0.7, label=label, color=colors.get(label, "gray"), edgecolor="white")

    plt.axvline(x=threshold, linestyle="--", color="black", linewidth=1.5, label=f"Threshold: {threshold}")
    
    plt.xlabel(f"Maximum Tanimoto Coefficient ({fp_type})", fontsize=16)
    plt.ylabel("Count", fontsize=16)
    plt.xticks(np.linspace(0, 1.0, 11))
    plt.title(f"Max Tanimoto Distribution ({fp_type.capitalize()} Fingerprint)", fontsize=16, pad=12)
    plt.legend(loc="upper right", fontsize=16)
    plt.grid(axis="y", alpha=0.3)
    plt.tight_layout()
    plt.show()

plt.rcParams.update({
    'font.size': 20,           # 设置全局字体大小
    'font.family': 'Arial',    # 设置字体为 Arial
    'axes.titlesize': 16,      # 坐标轴标题字体大小
    'axes.labelsize': 16,      # 坐标轴标签字体大小
    'xtick.labelsize': 16,     # X轴刻度字体大小
    'ytick.labelsize': 16,     # Y轴刻度字体大小
    'legend.fontsize': 16,     # 图例字体大小
    'figure.titlesize': 16     # 图形标题字体大小
})

# 设置坐标轴刻度向内
plt.rcParams.update({
    'xtick.direction': 'in',   # X轴刻度向内
    'ytick.direction': 'in',   # Y轴刻度向内
    'xtick.top': False,         # X轴刻度线显示在顶部
    'ytick.right': False        # Y轴刻度线显示在右侧
})
# 绘制直方图
plot_histogram(
    {"Optimized 70": opt_rdkit_70, "Optimized 80": opt_rdkit_80, "Generated": gen_rdkit, "Optimized 90": opt_rdkit_90},
    threshold=0.7, fp_type="rdkit"
)

plot_histogram(
    {"Optimized 70": opt_morgan_70, "Optimized 80": opt_morgan_80, "Generated": gen_morgan, "Optimized 90": opt_morgan_90},
    threshold=0.5, fp_type="morgan"
)
```

# 常规属性分布

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from rdkit import Chem
from rdkit.Chem import Descriptors
from rdkit.Chem import Crippen
import sascorer  # 确保你有这个文件
from tqdm import tqdm

tqdm.pandas()

# 1. 读取数据
train_df = pd.read_csv("Modeldata0406.csv")
gen_df = pd.read_excel("P-test.xlsx")
opt70_df = pd.read_csv("70_80similar_newSMILES0501.csv")
opt80_df = pd.read_csv("80_90similar_newSMILES0501.csv")
opt90_df = pd.read_csv("90_100similar_newSMILES0501.csv")

# 2. 提取 SMILES 列
train_smiles = train_df["SMILES"].dropna().unique().tolist()
gen_smiles = gen_df["SMILES"].dropna().unique().tolist()
opt70_smiles = opt70_df["SMILES"].dropna().unique().tolist()
opt80_smiles = opt80_df["SMILES"].dropna().unique().tolist()
opt90_smiles = opt90_df["SMILES"].dropna().unique().tolist()

# 3. 随机抽样训练集（比如抽5000个分子）
np.random.seed(42)
sample_size = 5000
if len(train_smiles) > sample_size:
    train_smiles = np.random.choice(train_smiles, sample_size, replace=False).tolist()

# 4. 定义计算属性的函数
def get_molecule_properties(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return np.nan, np.nan, np.nan, np.nan
    mol_weight = Descriptors.MolWt(mol)
    logp = Crippen.MolLogP(mol)
    tpsa = Descriptors.TPSA(mol)
    sa_score = sascorer.calculateScore(mol)
    return mol_weight, logp, tpsa, sa_score

# 5. 计算属性
train_properties = [get_molecule_properties(s) for s in tqdm(train_smiles, desc="Processing Training SMILES")]
gen_properties = [get_molecule_properties(s) for s in tqdm(gen_smiles, desc="Processing Generated SMILES")]
opt70_properties = [get_molecule_properties(s) for s in tqdm(opt70_smiles, desc="Processing Optimized 70 SMILES")]
opt80_properties = [get_molecule_properties(s) for s in tqdm(opt80_smiles, desc="Processing Optimized 80 SMILES")]
opt90_properties = [get_molecule_properties(s) for s in tqdm(opt90_smiles, desc="Processing Optimized 90 SMILES")]

# 6. 转为DataFrame
train_df_properties = pd.DataFrame(train_properties, columns=["MolWt", "LogP", "TPSA", "SA"])
gen_df_properties = pd.DataFrame(gen_properties, columns=["MolWt", "LogP", "TPSA", "SA"])
opt70_df_properties = pd.DataFrame(opt70_properties, columns=["MolWt", "LogP", "TPSA", "SA"])
opt80_df_properties = pd.DataFrame(opt80_properties, columns=["MolWt", "LogP", "TPSA", "SA"])
opt90_df_properties = pd.DataFrame(opt90_properties, columns=["MolWt", "LogP", "TPSA", "SA"])

```
