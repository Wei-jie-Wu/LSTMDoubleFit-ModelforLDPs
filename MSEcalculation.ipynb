```python
import pandas as pd
import numpy as np

data = pd.read_csv('AccuracyTest.csv')
data
```

```python
y_true = data.iloc[:, 6]
y_true
```

```python
y_pred = data.iloc[:, 7]
y_pred
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)
```

```python
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
```
